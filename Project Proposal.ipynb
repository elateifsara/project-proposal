```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**LinkedIn top 10 skills**

```python
linkedin_data = pd.read_excel('LinkedInTopSkills.xlsx')
```

```python
linkedin_data.head()
```

```python
groupedSkills = linkedin_data[['Skill','Rank']].groupby('Skill')
```

```python
groupedSkills.Rank.sum().sort_values(ascending=False).plot.bar()
```

**Notes**: This means that Mobile Development is the frequent skill followed by Network Information security 

**Jobs available in NYC**

```python
nycJobs_data = pd.read_csv('NYC_Jobs.csv')
```

```python
nycJobs_data.head()
```

```python
groupedJobs = nycJobs_data[['Business Title','# Of Positions']].groupby('Business Title')
```

```python
groupedJobs['# Of Positions'].count().sort_values(ascending=False).plot.bar()
```
